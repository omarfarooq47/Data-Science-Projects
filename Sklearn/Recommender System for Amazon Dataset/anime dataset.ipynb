### Joining the ratings with anime_id

```python
merged = pd.merge(tojoin,rating,how='inner',on='anime_id')
```

```python
merged.head()
```

```python
merged.loc[merged['rating']==-1,'rating']=0
#Setting all unknown ratings as 0
```

```python
from sklearn.metrics import mean_squared_error
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
#Calculating rmse function
```

```python
pivot = merged.pivot_table(index=['user_id'], columns=['name'], values='rating')
#User_item table
```

```python
pivot
```

```python
item = pivot.T
#Item-user table
```

```python
item
#For item-item collaborative filtering
```

```python
item.fillna(0,inplace=True)
#Replacing any NAN with 0
```

```python
item=item.loc[:,(item!=0).any(axis=0)]
#Removing users with 0 ratings
```

```python
itemsparse = sp.sparse.csr_matrix(item.values)
#Making the matrix sparse
```

```python
print(itemsparse)
print(itemsparse.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
item_similarity = cosine_similarity(itemsparse)
#Item similarity calculation
```

```python
item_similarity
```

```python
item_sim_df = pd.DataFrame(item_similarity, index = item.index, columns = item.index)
```

```python
item_sim_df
```

```python
#To find similar anime
def top10sim(anime_name):
    count = 1
    print('Similar shows to {} include:\n'.format(anime_name))
    for item in item_sim_df.sort_values(by = anime_name, ascending = False).index[1:11]:
        print('No. {}: {}'.format(count, item))
        count +=1
```

```python
top10sim('s.CRY.ed') #Checking
```

```python
#Weighed mean based prediction from item to item collaborative by weighed mean
def predictrating (user,anime):
    topanime = item_sim_df.sort_values(by = anime, ascending = False).index[1:11]
    ratinglist= []
    weightlist= []
    for i in topanime:
        if item.loc[i,user]!=0:
            ratinglist.append(item_sim_df.loc[anime,i] * item.loc[i,user]) #Weight(Similarity) * rating
            weightlist.append(item_sim_df.loc[anime,i]) #Weight (similarity)
    if len(weightlist)!=0:
        return sum(ratinglist)/sum(weightlist)
    else:
        return 0 #If no common anime found
```

```python
true=item.loc[item[3]>0][3] #Getting true ratings of anime watched by 3
```

```python
predicted = []
```

```python
topredict = true.index #Getting anime name that 3 watched
```

```python
for i in topredict:
    predicted.append(predictrating(3,i)) #Predicting value of watched anime
rmse(true,predicted)#Comparing result for 3
```

```python
true2=item.loc[item[5]>0][5] #Doing the same for person 2
predicted2 = []
topredict2 = true2.index
for i in topredict2:
    predicted2.append(predictrating(5,i))
rmse(true2,predicted2)
```

```python
totalrmse = [] #Calculating RMSE for the first 100 people
count = 0;
for i in item:
    truea=item.loc[item[i]>0][i]
    predicteda = []
    topredicta = truea.index
    for j in topredicta:
        predicteda.append(predictrating(i,j))
    totalrmse.append(rmse(truea,predicteda))
    count = count+1
    if count ==100:
        break
```

```python
print("Mean RMSE for 100 values")
print(sum(totalrmse)/100)
```
